```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
null = ["Nan",np.nan,"-","na"]
df = pd.read_csv("Basketball.csv", na_values=null)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.dropna(how="all")
```

```python
df = df.interpolate()
df.head()
```

```python
df.head()
```

```python
sns.pairplot(df[['PlayedGames','WonGames','LostGames']])
```

```python
sns.distplot(df['WonGames'])
```

```python
df.corr()
```

1).From the above correlation table we can conculde that Basket scored - won games and Basket Given - Lost games are highly correlated.  

2).When we compare the data with tournament champion and runner up, we conclude that there is less correlation between the them with the games won, lost or draw. So there is a possibility that tournament champion might have won or lost less games as compared to runner up.

3). Highest position held is giving negative correlation so it can be dropped.

4).Score of runner up(0.76) is more than the score of tournament champion(0.693417), which means score also does not make that much impact on the predictions. 

```python
df.drop(['Score','TournamentChampion','Runner-up','HighestPositionHeld'], axis=1, inplace=True)
```
